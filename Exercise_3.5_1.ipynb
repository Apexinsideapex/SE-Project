### 1. A function that takes a number and returns a list of its all possible factors (e.g., for 10, these are [1, 2, 5, 10]). Name this function "factors".

```python
#Defining a function that takes an integer as an input and returns a list of all its factors
def factors(num:int):
    factors_list = []
    for fact in range(1, num):
        if num%fact == 0:
            factors_list.append(fact)
    
    #Adding the number itself as a factor
    factors_list.append(num)
    
    return(factors_list)
```

```python
number = int(input("Enter a number: "))

#Calling the factors function
factors_list = factors(number)

print(f"The factors of {number} are {factors_list}")
```

### 2. A function that takes a number and returns True if the number is prime and False if the number is not prime (reuse the previous function!). E.g., for 7 it should return True, for 8 it should return False. Name this function "is_prime".

```python
#Defining a function that takes an integer as an input and checks if it is a prime number or not
def is_prime(number):
    #calling the factors function to get a list of factors
    factors_list = factors(number)
    
    #Checking if the number is prime
    #Number is prime if it has only two factors - 1 and the number itself
    print("Number is Prime" if len(factors_list) == 2 else "Number is not Prime")
        
    return
```

```python
number = int(input("Enter a number: "))

#Calling the is_prime function
is_prime(number)
```

### 3. A function that takes two numbers and returns their Highest Common Factor (e.g., for 12 and 18, it is 6). Name this function "hcf".

```python
#Defining a function that takes 2 integers as input and returns the list of common factors
def hcf(num1, num2):
    #Getting factors of num1 and num2
    factors1 = factors(num1)
    factors2 = factors(num2)
    common_factors = []
    
    #Checking for common factors
    for facts1 in factors1:
        if facts1 in factors2:
            common_factors.append(facts1)
            
    return common_factors
```

```python
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))

#Calling hcf function
common_factors = hcf(number1, number2)

#Last number is the HCF
print("The HCF is: ", common_factors[-1])
```
